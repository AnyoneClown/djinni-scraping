```python
import ast

import pandas as pd
import matplotlib.pyplot as plt

from collections import Counter

%matplotlib inline
```

# Data extraction

```python
vacancies = pd.read_csv("../scraping/vacancies.csv")
vacancies.head()
```

# Data cleaning

```python
vacancies = vacancies.replace(r"\n"," ", regex=True)
vacancies = vacancies.replace(r"\s+", " ", regex=True)
vacancies.head(10)
```

# Data wrangling

```python
vacancies["requirements"] = vacancies["requirements"].apply(ast.literal_eval)

technology_counts = Counter([tech for sublist in vacancies["requirements"] for tech in sublist])
sorted_technologies = technology_counts.most_common(20)
techs, counts = zip(*sorted_technologies)
```

# Data Vizualization

```python
plt.figure(figsize=(14, 6))
plt.bar(techs, counts, color="skyblue")
plt.title("Top 20 Technology Occurrence Frequency")
plt.xlabel("Technology")
plt.ylabel("Occurrence count")
plt.xticks(rotation=45, ha="right")
plt.savefig("plots/technology_occurrence.png")
plt.show()
```
